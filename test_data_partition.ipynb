```python
import pandas as pd
import random
from sklearn.model_selection import train_test_split
```

```python
seed = 42
```

```python
df  = pd.read_csv('./data/pacs_v1.0/centralize.csv')
```

```python
df
```

```python
print(len(df[df['split'] == 'train']))
print(len(df[df['split'] == 'val']))
print(len(df[df['split'] == 'test']))
```

```python
domain_list = df['domain'].unique()
category_list = df['category'].unique()
```

```python
for domain in domain_list:
    for category in category_list:
        category_df = df[(df['category'] == category) & (df['domain'] == domain)]
        category_length = len(category_df)
        category_index = category_df.index.values.tolist()
        train_idx, test_idx = train_test_split(category_index, train_size=0.9, random_state=seed)
        train_idx, val_idx = train_test_split(train_idx, train_size=0.8889, random_state=seed)
        df.loc[train_idx, 'split'] = 'train'
        df.loc[val_idx, 'split'] = 'val'
        df.loc[test_idx, 'split'] = 'test'
```

```python
df
```

```python
df.to_csv('./data/pacs_v1.0/centralize.csv', sep=',', index=False, encoding='utf-8')
```
