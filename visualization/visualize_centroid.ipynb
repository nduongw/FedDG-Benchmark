```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
```

```python
centroid_list = []
for round in range(50):
    centroid = np.load(f'./centroid_at_round{round}.npy')
    centroid_list.append(np.transpose(centroid))
```

```python
centroid_list[0].shape
```

```python
# round = 19
for round in range(50):
    similarity_matrix = cosine_similarity(centroid_list[round])
    sns.heatmap(data=similarity_matrix, annot=True, cmap="YlGnBu")
    plt.show()
```
