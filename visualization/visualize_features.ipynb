```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
```

```python
data_list = []
for i in range(14):
    features = np.load(f'./results/data_round{i}_id_val.npz', allow_pickle=True)
    data_list.append(features)
```

```python
feats_list = []
labels_list = []
for val in data_list:
    data = val['data']
    feats = np.concatenate(data.item()['data'], axis=0)
    labels = np.concatenate(data.item()['label'], axis=0)
    feats_list.append(feats)
    labels_list.append(labels)
    
```

```python
for i in range(20):
    tsne = TSNE(n_components=2)
    post_feats = tsne.fit_transform(feats_list[i])
    plt.scatter(post_feats[:, 0], post_feats[:, 1], c=labels_list[i])
    plt.show()
```
